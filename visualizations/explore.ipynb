```python
import json
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d

# Load the evaluation scores
with open('eval_scores.json', 'r') as f:
    eval_scores = json.load(f)

# Extract data
epochs = [entry['epoch'] for entry in eval_scores]
train_loss = [entry['train_loss'] for entry in eval_scores]
bleu_1 = [entry['bleu_1'] for entry in eval_scores]
bleu_2 = [entry['bleu_2'] for entry in eval_scores]
bleu_3 = [entry['bleu_3'] for entry in eval_scores]
bleu_4 = [entry['bleu_4'] for entry in eval_scores]
meteor = [entry['meteor'] for entry in eval_scores]
time = [entry['time'] for entry in eval_scores]

# Apply smoothing with a sigma of 2
train_loss_smooth = gaussian_filter1d(train_loss, sigma=2)
bleu_1_smooth = gaussian_filter1d(bleu_1, sigma=2)
bleu_2_smooth = gaussian_filter1d(bleu_2, sigma=2)
bleu_3_smooth = gaussian_filter1d(bleu_3, sigma=2)
bleu_4_smooth = gaussian_filter1d(bleu_4, sigma=2)
meteor_smooth = gaussian_filter1d(meteor, sigma=2)
time_smooth = gaussian_filter1d(time, sigma=2)

# Create a figure with subplots
fig, axes = plt.subplots(4, 2, figsize=(15, 15))
fig.suptitle('Model Evaluation Metrics Over Epochs (Smoothed)', fontsize=16)

# Plot training loss
axes[0, 0].plot(epochs, train_loss, 'b-', alpha=0.3, label='Raw')
axes[0, 0].plot(epochs, train_loss_smooth, 'b-', label='Smoothed')
axes[0, 0].set_title('Training Loss')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('Loss')
axes[0, 0].grid(True)
axes[0, 0].legend()

# Plot BLEU-1
axes[0, 1].plot(epochs, bleu_1, 'g-', alpha=0.3, label='Raw')
axes[0, 1].plot(epochs, bleu_1_smooth, 'g-', label='Smoothed')
axes[0, 1].set_title('BLEU-1 Score')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Score')
axes[0, 1].grid(True)
axes[0, 1].legend()

# Plot BLEU-2
axes[1, 0].plot(epochs, bleu_2, 'r-', alpha=0.3, label='Raw')
axes[1, 0].plot(epochs, bleu_2_smooth, 'r-', label='Smoothed')
axes[1, 0].set_title('BLEU-2 Score')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Score')
axes[1, 0].grid(True)
axes[1, 0].legend()

# Plot BLEU-3
axes[1, 1].plot(epochs, bleu_3, 'c-', alpha=0.3, label='Raw')
axes[1, 1].plot(epochs, bleu_3_smooth, 'c-', label='Smoothed')
axes[1, 1].set_title('BLEU-3 Score')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Score')
axes[1, 1].grid(True)
axes[1, 1].legend()

# Plot BLEU-4
axes[2, 0].plot(epochs, bleu_4, 'm-', alpha=0.3, label='Raw')
axes[2, 0].plot(epochs, bleu_4_smooth, 'm-', label='Smoothed')
axes[2, 0].set_title('BLEU-4 Score')
axes[2, 0].set_xlabel('Epoch')
axes[2, 0].set_ylabel('Score')
axes[2, 0].grid(True)
axes[2, 0].legend()

# Plot METEOR
axes[2, 1].plot(epochs, meteor, 'y-', alpha=0.3, label='Raw')
axes[2, 1].plot(epochs, meteor_smooth, 'y-', label='Smoothed')
axes[2, 1].set_title('METEOR Score')
axes[2, 1].set_xlabel('Epoch')
axes[2, 1].set_ylabel('Score')
axes[2, 1].grid(True)
axes[2, 1].legend()

# Plot Time
axes[3, 0].plot(epochs, time, 'k-', alpha=0.3, label='Raw')
axes[3, 0].plot(epochs, time_smooth, 'k-', label='Smoothed')
axes[3, 0].set_title('Evaluation Time')
axes[3, 0].set_xlabel('Epoch')
axes[3, 0].set_ylabel('Time (seconds)')
axes[3, 0].grid(True)
axes[3, 0].legend()

# Remove the last subplot (empty)
fig.delaxes(axes[3, 1])

# Adjust layout and save
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Load the evaluation scores
with open('eval_scores.json', 'r') as f:
    eval_scores = json.load(f)

# Extract data and filter for every 5 epochs
n = 1
epochs = [entry['epoch'] for entry in eval_scores if entry['epoch'] % n == 0]
train_loss = [entry['train_loss'] for entry in eval_scores if entry['epoch'] % n == 0]
bleu_1 = [entry['bleu_1'] for entry in eval_scores if entry['epoch'] % n == 0]
bleu_2 = [entry['bleu_2'] for entry in eval_scores if entry['epoch'] % n == 0]
bleu_3 = [entry['bleu_3'] for entry in eval_scores if entry['epoch'] % n == 0]
bleu_4 = [entry['bleu_4'] for entry in eval_scores if entry['epoch'] % n == 0]
meteor = [entry['meteor'] for entry in eval_scores if entry['epoch'] % n == 0]
time = [entry['time'] for entry in eval_scores if entry['epoch'] % n == 0]

# Plot training loss
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_loss, 'b-o')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Cross Entropy Loss')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot BLEU-1
plt.figure(figsize=(10, 6))
plt.plot(epochs, bleu_1, 'g-o')
plt.title('BLEU-1 Score')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot BLEU-2
plt.figure(figsize=(10, 6))
plt.plot(epochs, bleu_2, 'r-o')
plt.title('BLEU-2 Score')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot BLEU-3
plt.figure(figsize=(10, 6))
plt.plot(epochs, bleu_3, 'c-o')
plt.title('BLEU-3 Score')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot BLEU-4
plt.figure(figsize=(10, 6))
plt.plot(epochs, bleu_4, 'm-o')
plt.title('BLEU-4 Score')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot METEOR
plt.figure(figsize=(10, 6))
plt.plot(epochs, meteor, 'y-o')
plt.title('METEOR Score')
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot Time
plt.figure(figsize=(10, 6))
plt.plot(epochs, time, 'k-o')
plt.title('Evaluation Time Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot BLEU-1 and METEOR scores on different axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# First axis (left) for BLEU-1
ax1.plot(epochs, bleu_1, 'r-o', label='BLEU-1')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('BLEU-1 Score', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Create second axis (right) for METEOR
ax2 = ax1.twinx()
ax2.plot(epochs, meteor, 'b-o', label='METEOR')
ax2.set_ylabel('METEOR Score', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Add title and grid
plt.title('BLEU-1 and METEOR Scores')
ax1.grid(True)

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

plt.tight_layout()
plt.show()
```
